```python
import warnings
warnings.filterwarnings("ignore")
```

```python
import pandas as pd
import numpy as np
df = pd.read_csv('data/지역별 환경 교과목 자격증 소유 교원수.csv', encoding='utf-8')
df
```

```python
df['환경 교과목 교원 수'] = df['환경 교과목 교원 수'].replace('-', 0)
df
```

```python
df['환경 교과목 교원 수'] = pd.to_numeric(df['환경 교과목 교원 수'])
df.info()
```

```python
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

font_location = 'C:/Windows/Fonts/NanumGothic.ttf'
font_name = fm.FontProperties(fname=font_location).get_name()
plt.rc('font', family=font_name)
```

```python
plt.figure(figsize = (15, 9))
plt.plot(df['지역'], df['환경 교과목 교원 수'], marker='o', color='#B088F9')
plt.title('지역별 환경 교과목 자격증을 소유한 교원 수', fontsize=25)
plt.xticks(fontsize=16)
plt.yticks(np.arange(0, 21, 2), fontsize=16)
plt.xlabel('지역 교육청', fontsize=15)
plt.ylabel('교원 수', fontsize=15)
plt.show()
```

```python
df2 = pd.read_csv('data/서울특별시 과목별 자격증 보유 교직원 현황.csv', encoding='utf-8')
df2
```

```python
df2['교원 수'] = df2['교원 수'].replace('-', 0)
df2['교원 수'] = pd.to_numeric(df2['교원 수'])
df2.info()
```

```python
plt.figure(figsize=(18, 9))
plt.plot(df2['과목명'], df2['교원 수'], marker='o', color='#B088F9')
plt.title('서울특별시교육청 과목별 자격증 보유 현황', fontsize=25)
plt.xticks(fontsize=16, rotation=70)
plt.yticks(fontsize=16)
plt.xlabel('과목명', fontsize=15)
plt.ylabel('교원 수', fontsize=15)
plt.show()
```
