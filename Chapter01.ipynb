## 1.1 데이터를 지식으로 바꾸는 지능적인 시스템 구축

데이터에서 지식을 추출하여 예측하는 자기 학습 알고리즘과 관련된 인공지능의 하위 분야로 머신 러닝이 출현. 

사람이 수동으로 대량의 데이터를 분석하여 규칙을 유도하고 모델을 만드는 대신, 머신 러닝이 데이터에서 더 효율적으로 지식을 추출해 예측 모델과 데이터 기반의 의사 결정 성능을 점진적으로 향상. 

## 1.2 머신 러닝의 세 가지 종류

< 머신러닝의 세 가지 학습 종류 >
1. 지도학습
    * 레이블된 데이터
    * 직접 피드백
    * 출력 및 미래 예측
    
2. 비지도 학습
    * 레이블 및 타깃 없음
    * 피드백 없음
    * 데이터에서 숨겨진 구조 찾기
  
3. 강화 학습
    * 결정 과정
    * 보상 시스템
    * 연속된 행동에서 학습
    
### 1.2.1 지도 학습으로 미래 예측

지도 학습의 목적 : 레이블된 훈련 데이터에서 모델을 학습하여 본 적 없는 미래 데이터에 대한 예측을 만드는 것. 

**지도** 란? 희망하는 출력 신호 즉 레이블이 있는 일련의 샘플 의미.

예시) 스팸 이메일 필터링

레이블된 이메일 데이터셋에서 지도 학습 머신 러닝 알고리즘을 사용해 모델을 훈련할 수 있다. 데이터셋은 스팸 또는 스팸이 아닌 이메일로 정확하게 표시되어 있어 이를 훈련한 모델은 새로운 이메일이 두 개의 범주 중 어디에 속하는지 예측. 

지도 학습은 예시처럼 개별 클래스 레이블이 있는 지도 학습을 **분류(Classification)** , 연속적인 값을 출력하는 **회귀(Regression)** 의 두 가지 종류가 있다.

#### **분류 : 클래스 레이블 예측** 

* 목적 : 과거의 관측을 기반으로 새로운 샘플의 범주형 클래스 레이블 예측
* 이진 분류(binary classification) : 두 개의 클래스 사이 구분 (ex. 스팸 이메일 감지 등)
* 다중 분류(multiclass classification) : 두 개 이상의 클래스 레이블을 가진 경우 (ex. 손으로 쓴 글자 인식)
* 결정 경계 : 지도 학습 알고리즘을 사용해 두 클래스를 구분할 수 있는 규칙을 학습해 선으로 나타낸 것

#### **회귀 : 연속적인 출력 값 예측**

* 목적 : 예측 변수(설명변수, 입력)와 연속적인 반응 변수(출력, 타깃)가 주어졌을 때 출력 값을 예측하는 두 변수 사이의 관계 파악  

### 1.2.2 강화 학습으로 반응형 문제 해결

* **강화 학습**의 목적 : 환경과 상호 작용하여 시스템(에이전트) 성능을 향상하는 것
* 강화 학습의 피드백 : 정답 레이블이나 값이 아닌 보상함수(얼마나 행동이 좋은지를 측정한 값)
* 에이전트(agent) : 환경화 상호 작용하여 보상이 최대화되는 일련의 행동을 강화 학습으로 학습

### 1.2.3 비지도 학습으로 숨겨진 구조 발견

지도 학습에서는 모델을 훈련할 때 사전에 옳은 답을 알고 있고, 강화 학습에서는 에이전트의 특정 행동을 어떻게 보상할지 그 측정 방법 정의.

**비지도 학습**에서는 레이블되지 않거나 구조를 알 수 없는 데이터를 다뤄 알려진 출력 값이나 보상 함수의 도움을 받지 않고 의미 있는 정보를 추출하기 위해 데이터 구조 탐색

#### **군집 : 서브그룹 찾기**

**군집(clustering)** : 사전 정보 없이 쌓여 있는 그룹 정보를 의미 있는 서브그룹 또는 클러스터로 조직하는 탐색적 데이터 분석 기법

분석 과정에서 만들어진 각 클러스터는 어느 정도의 유사성을 공유하고 다른 클러스터와는 비슷하지 않은 샘플 그룹을 형성하므로 비지도 분류라고 함.

#### **차원 축소 : 데이터 압축**

고차원의 데이터를 다룰 때 즉, 하나의 관측 샘플에 많은 측정 지표가 존재할 때 머신 러닝 알고리즘의 계산 성능과 저장 공간의 한계가 있을 수 있다. 이때 **비지도 차원 축소**는 잡음 데이터를 제거하기 위해 특성 전처리 단계에서 적용하는 방법이다. 

잡음 데이터는 특정 알고리즘의 예측 성능을 감소시킬 수 있어 차원 축소를 이용해 관련 있는 정보를 유지하면서 더 작은 차원의 부분 공간으로 데이터를 압축한다. 

## 1.3 기본 용어와 표기법 소개

머신 러닝의 예제 : 붓꽃(iris) 데이터셋

붓꽃 데이터셋은 Setosa, Versicolor, Virginica 세 종류의 150개의 붓꽃 샘플을 가진다. 각 붓꽃 샘플은 하나의 행으로 표현되어 센티미터 단위의 측정 값은 열에 저장되어 특성이라고도 한다. 

## 1.4 머신 러닝 시스템 구축 로드맵

### 1.4.1 전처리 : 데이터 형태 갖추기

주어진 원본 데이터의 형태와 모습이 학습 알고리즘이 최적의 성능을 내기에 적합한 경우는 매우 드물다. 따라서 데이터 전처리 단계는 모든 머신 러닝 애플리케이션 단계에서 가장 중요하다.

붓꽃 데이터셋을 예시로 하자면 원본 데이터는 일련의 꽃 이미지들이고, 여기서 의미 있는 특성을 추출해야 한다. 이때 최적의 성능을 내기 위해서는 선택된 특성을 [0, 1] 범위로 변환하거나 평균이 0이고 단위 분산을 가진 표준 정규 분포로 변환하는 등의 작업을 통해 같은 스케일을 가져야 한다. 또 일부 특성은 상관관계가 높아 중복된 정보를 가질 수 있어 차원 축소 기법을 이용해 특성을 저차원 부분 공간으로 압축해 저장 공간이 덜 필요하고 학습 알고리즘을 빨리 실행할 수 있도록 한다. 

그리고 머신 러닝 알고리즘이 훈련 데이터셋에서 잘 작동하고 새로운 데이터에서도 잘 일반화되는지 확인하기 위해 데이터셋을 랜덤하게 훈련 세트와 테스트 세트로 나눠 훈련 세트를 통해 머신 러닝 모델을 훈련 및 최적화 하고, 테스트 세트를 이용해 최종 모델을 평가한다.

### 1.4.2 예측 모델 훈련과 선택

가장 좋은 모델을 훈련하고 선택하기 위해서는 성능을 측정할 지표를 결정해야 한다.

* 정확도(Accuracy) : 정확히 분류된 샘플 비율

* 교차 검증 기법 : 모델의 일반화 성능을 예측하기 위해 훈련 데이터를 훈련 세트와 검증 세트로 더 나누는 방법

### 1.4.3 모델을 평가하고 본 적 없는 샘플로 예측

훈련 세트에서 최적의 모델을 선택한 후 테스트 세트를 이용해 이전에 본 적 없는 데이터에서 얼마나 성능을 내는지 예측하여 일반화 오차를 예상. 

단, 특성 스케일 조정과 차원 축소는 훈련 세트에만 사용해 얻은 것.
