## 3.2 사이킷런 첫걸음 : 퍼셉트론 훈련

#### **붓꽃 데이터 불러오기**

150개 꽃 샘플에서 꽃잎 길이와 꽃잎 너비를 특성 행렬 X에 할당하고 이에 상응하는 꽃 품종에 해당하는 클래스 레이블을 벡터 y에 할당

시각화를 위해 붓꽃 데이터셋에서 두 개의 특성만 사용.

```python
from sklearn import datasets
import numpy as np

iris = datasets.load_iris()
X = iris.data[:, [2, 3]]
y = iris.target
# np.unique(y) : iris.target에 저장된 세 개의 고유한 클래스 레이블 반환
print('클래스 레이블:', np.unique(y))   # 붓꽃 클래스 이름은 Iris-setosa, Iris-versicolor, Iris-virginica 여기서는 0, 1, 2로 표현
```

#### **데이터셋 나누기**

훈련 세트와 테스트 세트로 분할

```python
from sklearn.model_selection import train_test_split    # train_test_split 함수 : 분할하기 전에 데이터셋을 먼저 섞는다.

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1, stratify=y)    
# random_state 매개변수로 고정된 랜덤 시드 전달(random_state = 1) : 실행 결과 재현 가능
# stratify : 계층화 기능 사용(train_test_split 함수가 훈련 세트와 테스트 세트의 클래스 레이블 비율을 입력 데이터셋과 동일하게 만든다는 의미)
```

numpy bincount 함수(배열에 있는 고유한 값의 등장횟수 계산)를 이용한 계층화가 잘 되었는지 확인

```python
print('y의 레이블 카운트:', np.bincount(y))
print('y_train의 레이블 카운트:', np.bincount(y_train))
print('y_test의 레이블 카운트:', np.bincount(y_test))
```

사이킷런의 preprocessing 모듈의 StandardScaler 클래스 사용한 **특성 표준화**

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()   # 새로운 StandardScalar 객체를 sc 변수에 할당
sc.fit(X_train)         # fit 메서드 : 훈련 세트와 각 특성 차원마다 샘플 평균과 표준편차 계산
# 훈련 세트와 테스트 세트의 샘플이 서로 같은 비율로 이동되도록 표준화
X_train_std = sc.transform(X_train)   # transform 메서드 : 계산된 샘플 평균과 표준편차를 사용해 훈련 세트 표준화
X_test_std = sc.transform(X_test)     # transform 메서드 : 계산된 샘플 평균과 표준편차를 사용해 테스트 세트 표준화
```

#### **퍼셉트론 모델 훈련**

세 개의 붓꽃 클래스를 퍼셉트론에 주입

```python
from sklearn.linear_model import Perceptron   # linear_model 모듈에서 Perceptron 클래스 로드

# 새로운 Perceptron 객체 생성
ppn = Perceptron(max_iter=40, eta0=0.1, tol=1e-3, random_state=1)   # eta0 : 학습률 eta, max_iter : 훈련 세트 반복할 에포크 횟수, random_state : 훈련 세트를 섞은 결과 재현
ppn.fit(X_train_std, y_train)       # fit 메서드를 사용해 모델 훈련
```

**precict 메서드로 예측**

```python
y_pred = ppn.predict(X_test_std)
print('잘못 분류된 샘플 개수: %d' % (y_test != y_pred).sum())   # 45개 샘플에서 1개 잘못 분류 -> 오차 : 1/45 = 약0.022
```

**분류 정확도 계산**

테스트 세트에서 퍼셉트론의 분류 정확도

```python
from sklearn.metrics import accuracy_score    # metrics 모듈에 다양한 성능 지표 구현해놓음

print('정확도: %.2f' % accuracy_score(y_test, y_pred))    # y_test : 진짜 클래스 레이블, y_pred : 예측한 클래스 레이블
```

score 메서드 : 분류기의 예측 정확도 계산

predict 메서드와 accuracy_score 메서드를 연결해 정확도 계산

```python
print('정확도: %.2f' % ppn.score(X_test_std, y_test))
```

**새로운 모델의 결정 정계 시각화**

plot_decision_regions 함수를 사용해 세 개의 샘플을 잘 구분하는지 시각화(테스트 세트와 샘플을 작은 원으로 표시)

```python
from matplotlib.colors import ListedColormap
import matplotlib.pyplot as plt

def plot_decision_regions(X, y, classifier, test_idx=None, resolution=0.02):
  # 마커와 컬러맵 설정
  markers = ('s', 'x', '^', 'v')
  colors = ('red', 'blue', 'lightgreen', 'gray', 'cyan')
  cmap = ListedColormap(colors[:len(np.unique(y))])

  # 결정 정계 표시
  x1_min, x1_max = X[:, 0].min() - 1, X[:, 0].max() + 1
  x2_min, x2_max = X[:, 1].min() - 1, X[:, 1].max() + 1
  xx1, xx2 = np.meshgrid(np.arange(x1_min, x1_max, resolution),
                         np.arange(x2_min, x2_max, resolution))
  Z = classifier.predict(np.array([xx1.ravel(), xx2.ravel()]).T)
  Z = Z.reshape(xx1.shape)
  plt.contourf(xx1, xx2, Z, alpha = 0.3, cmap = cmap)
  plt.xlim(xx1.min(), xx1.max())
  plt.ylim(xx2.min(), xx2.max())

  for idx, cl in enumerate(np.unique(y)):
    plt.scatter(x=X[y == cl, 0], y=X[y == cl, 1],
                alpha = 0.8, c = colors[idx],
                marker = markers[idx], label = cl,
                edgecolor = 'black')
    
  # 테스트 샘플을 부각하여 표시
  if test_idx:
    X_test, y_test = X[test_idx, :], y[test_idx]

    plt.scatter(X_test[:, 0], X_test[:, 1],
                facecolors = 'none', edgecolor = 'black', alpha = 1.0, 
                linewidth = 1, marker = 'o',
                s = 100, label = 'test set')
```

```python
X_combined_std = np.vstack((X_train_std, X_test_std))
y_combined = np.hstack((y_train, y_test))
plot_decision_regions(X = X_combined_std,
                      y = y_combined,
                      classifier = ppn,
                      test_idx = range(105, 150))
plt.xlabel('petal length [standardized')
plt.ylabel('petal width [standardized]')
plt.legend(loc = 'uppper left')
plt.tight_layout()
plt.show()
```

결론 : 세 개의 붓꽃 클래스는 선형 결정 정계로 완벽하게 분류하지 못한다.

퍼셉트론 알고리즘은 선형적으로 구분되지 않는 데이터셋에는 수렴하지 못해 퍼셉트론 알고리즘을 추천하지 않는다.

## 3.3 로지스틱 회귀를 사용한 클래스 확률 모델링

퍼셉트론 규칙의 단점 : 클래스가 선형적으로 구분되지 않을 때 수렴할 수 없다는 것.

### 3.3.1 로지스틱 회귀의 이해와 조건부 확률

* 오즈 : 특정 이벤트가 발생할 확률
* 오즈비 = P / (1-P)
* 로짓 함수 : 오즈비에 로그 함수를 취한 것
* 로지스틱 시그모이드 함수 : 어떤 샘플이 특정 클래스에 속할 확률을 예측하는 것에 관심 있으므로 로짓 함수를 뒤집은 함수

**시그모이드 곡선 시각화**

```python
import matplotlib.pyplot as plt
import numpy as np

def sigmoid(z):
  return 1.0 / (1.0 + np.exp(-z))
```

```python
z = np.arange(-7, 7, 0.1)
phi_z = sigmoid(z)
plt.plot(z, phi_z)
plt.axvline(0.0, color='k')
plt.ylim(-0.1, 1.1)
plt.xlabel('z')
plt.ylabel('$\phi (z)$')

# y축의 눈금과 격자선
plt.yticks([0.0, 0.5, 1.0])
ax = plt.gca()
ax.yaxis.grid(True)
plt.tight_layout()
plt.show()
```

z가 무한대로 가면 시그모이드 함수는 1에 가까워지고 -무한대로 가면 분모가 커져 0에 수렴한다.

따라서 시그모이드 함수는 0과 1 사이의 값으로 반환한다.

### 3.3.2 로지스틱 비용 함수의 가중치 학습

예시) 샘플이 하나인 경우 시그모이드 함수 값에 대한 분류 비용 그래프

**시그모이드 활성화 대비 로지스틱 비용 그래프**

```python
def cost_1(z):
   return -np.log(sigmoid(z))
  
def cost_0(z):
  return -np.log(1 - sigmoid(z))
```

```python
z = np.arange(-10, 10, 0.1)
phi_z = sigmoid(z)
c1 = [cost_1(x) for x in z]
plt.plot(phi_z, c1, label='J(w) if y=1')
c0 = [cost_0(x) for x in z]
plt.plot(phi_z, c0, linestyle='--', label='J(w) if y=0')
plt.ylim(0.0, 5.1)
plt.xlim([0, 1])
plt.xlabel('$\phi$(z)')
plt.ylabel('J(w)')
plt.legend(loc = 'best')
plt.tight_layout()
plt.show()
```

그래프 설명 :

x축은 0과 1까지 범위의 시그모이드 활성화 값, y축은 해당하는 로지스틱 비용

클래스 1에 속한 샘플을 정확히 예측하면 비용이 0에 가까워짐(실선)

비슷하게 클래스 0에 속한 샘플을 y=0으로 정확히 예측하면 y축의 비용이 0에 가까워짐(점선)

예측이 잘못되면 비용은 무한대, 즉 잘못된 예측에 점점 더 큰 비용 부여

### 3.3.3 아달린 구현을 로지스틱 회귀 알고리즘으로 변경

에포크마다 모든 훈련 샘플을 분류하는 비용을 계산.

선형 활성화 함수를 시그모이드 활성화로 바꾸고 임계 함수가 클래스 레이블 -1과 1이 아니고 0과 1을 반환하도록 변경.

```python
class LogisticRegressionGD(object):
  """ 경사 하강법을 사용한 로지스틱 회귀 분류기

  --매개변수--
  eta : 학습률
  n_iter : 훈련 데이터셋 반복 횟수
  random_state : 가중치 무작위 초기화를 위한 난수 생성기 시드

  --속성--
  w_ : 학습된 가중치
  cost_ : 에포크마다 누적된 로지스틱 비용 함수 값(리스트 형태)

  """

  def __init__(self, eta=0.05, n_iter=100, random_state=1):
    self.eta = eta
    self.n_iter = n_iter
    self.random_state = random_state
  def fit(self, X, y):
    """ 훈련 데이터 학습

    --매개변수--
    X : n_samples개의 샘플과 n_features개의 특성으로 이루어진 훈련 데이터 [n_samples, n_features]
    y : 타깃 값 [n_samples]

    --반환값--
    self : object

    """
    rgen = np.random.RandomState(self.random_state)
    self.w_ = rgen.normal(loc=0.0, scale=0.01,
                           size=1+X.shape[1])
    self.cost_ = []

    for i in range(self.n_iter):
      net_input = self.net_input(X)
      output = self.activation(net_input)
      errors = (y - output)
      self.w_[1:] += self.eta * X.T.dot(errors)
      self.w_[0] += self.eta * errors.sum()

      # 오차 제곱합 대신 로지스틱 비용 계산
      cost = (-y.dot(np.log(output)) - ((1-y).dot(np.log(1-output))))
      self.cost_.append(cost)
    return self

  def net_input(self, X):
    """ 최종 입력 계산 """
    return np.dot(X, self.w_[1:]) + self.w_[0]

  def activation(self, z):
    """ 로지스틱 시그모이드 활성화 계산 """
    return 1. / (1. + np.exp(-np.clip(z, -250, 250)))

  def predict(self, X):
    """ 단위 계단 함수를 사용하여 클래스 레이블을 반환합니다 """
    return np.where(self.net_input(X) >= 0.5, 1, 0)
    #다음과 동일함
    #return np.where(self.activation(self.net_input(X)) >= 0.5, 1, 0)
```

이진 분류 문제에만 적용 가능 -> Iris-setosa와 Iris-versicolor(클래스 0과 1)에 대해서 적용

```python
X_train_01_subset = X_train[(y_train == 0) | (y_train == 1)]
y_train_01_subset = y_train[((y_train == 0) | (y_train == 1))]
lrgd = LogisticRegressionGD(eta=0.05,
                            n_iter=1000, 
                            random_state=1)
lrgd.fit(X_train_01_subset,
         y_train_01_subset)
plot_decision_regions(X=X_train_01_subset,
                      y=y_train_01_subset,
                      classifier=lrgd)
plt.xlabel('petal length [standardized]')
plt.ylabel('petal width [standardized]')
plt.legend(loc='upper left')
plt.tight_layout()
plt.show()
```

### 3.3.5 규제를 사용하여 과대적합 피하기

**과대적합(Overfitting)** : 모델이 훈련 데이터로는 잘 동작하지만 본 적 없는 데이터(테스트 데이터)로는 잘 일반화되지 않는 현상 -> 모델 파라미터가 너무 많아 주어진 데이터에서 너무 복잡한 모델을 만들기 때문

**과소적합(Underfitting)** : 훈련 데이터에 있는 패턴을 감지할 정도로 충분히 모델이 복잡하지 않다는 것 의미. -> 새로운 데이터에서도 성능이 낮을 것.

좋은 편향-분산 트레이드오프를 찾는 방법은 규제를 사용해 모델의 복잡도를 조정하는 것

* 규제 : 공선성(특성 간의 높은 상관관계)을 다루거나 데이터에서 잡음을 제거해 과대적합을 방지할 수 있는 방법으로 과도한 파라미터(가중치) 값을 제한하기 위해 추가적인 정보(편향)을 주입하는 방법

규제 항은 모델 훈련 과정에서 가중치를 줄이는 역할 수행. 규제 하이퍼파라미터 lambda를 사용해 가중치를 작게 유지하면서 훈련 데이터에 얼마나 잘 맞출지를 조정. lambda 값을 증가하면 규제 강도가 높아짐.

사이킷런의 LogisticRegression 클래스의 매개변수 C가 규제 하이퍼파라미터의 역수로, 감소하면 규제 강도 증가.

**규제 매개변수에 따른 가중치 크기의 변화**

```python
weights, params = [], []
for c in np.arange(-5, 5):
  lr = LogisticRegression(solver='liblinear', multi_class='auto', C=10.**c,
                          random_state=1)
  lr.fit(X_train_std, y_train)
  weights.append(lr.coef_[1])
  params.append(10.**c)

weights = np.array(weights)
plt.plot(params, weights[:, 0],
         label='petal length')
plt.plot(params, weights[:, 1], linestyle='--',
         label='petal width')
plt.ylabel('weights coefficient')
plt.xlabel('C')
plt.legend(loc='uppper left')
plt.xscale('log')
plt.show()
```

코드 결과, 역 규제 매개변수 C의 값을 바꾸면서 열 개의 로지스텍 회귀 모델 훈련

시연을 위해 모든 분류기에서 클래스 1의 가중치 값만 사용한다. (여기서는 두 번째 클래스인 Iris-versicolor), 다중 분류에는 OvR 기법 사용

그래프를 통해 매개변수 C가 감소하면 가중치 절댓값이 줄어드는 것을 알 수 있다. 즉, 규제 강도가 증가한다.

## 3.5 커널 SVM을 사용하여 비선형 문제 풀기

### 3.5.1 선형적으로 구분되지 않는 데이터를 위한 커널 방법

**샘플 데이터셋 생성**

넘파이 logical_xor 함수를 사용해 XOR 형태의 간단한 데이터셋 생성

100개의 샘플은 클래스 레이블 1로 할당, 나머지 100개의 샘플은 클래스 레이블 -1로 할당

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(1)
X_xor = np.random.randn(200, 2)
y_xor = np.logical_xor(X_xor[:, 0] > 0,
                       X_xor[:, 1] > 0)
y_xor = np.where(y_xor, 1, -1)
plt.scatter(X_xor[y_xor == 1, 0],
            X_xor[y_xor == 1, 1],
            c='b', marker='x',
            label='1')
plt.scatter(X_xor[y_xor == -1, 0],
            X_xor[y_xor == -1, 1],
            c='r', marker='s',
            label='-1')
plt.xlim([-3, 3])
plt.ylim([-3, 3])
plt.legend(loc='best')
plt.tight_layout()
plt.show()
```

랜덤한 잡음이 섞인 XOR 데이터셋 생성.

양성 클래스와 음성 클래스를 선형 초평만으로 구분할 수 없다. 따라서 이렇게 선형적으로 구분되지 않는 데이터를 다루는 **커널방법** 필요.

* 커널 방법의 기본 아이디어 : 매핑 함수를 사용해 원본 특성의 비선형 조합을 선형적으로 구분되는 고차원 공간에 투영하는 것.

## 3.6 결정 트리 학습

**결정트리** : 훈련 데이터에 있는 특성을 기반으로 샘플의 클래스 레이블을 추정할 수 있는 일련의 질문 학습. 

결정 알고리즘을 사용하면 트리의 루트에서 시작해 정보 이득이 최대가 되는 특성으로 데이터를 나눈다. 반복 과정을 통해 리프 노드가 순수해질 때까지 모든 자식 노드에서 이 분할 작업을 반복. 즉, 이 노드의 샘플은 동일한 클래스에 속함. 

하지만 이렇게 하면 노드가 많은 깊은 트리가 만들어지고 과대적합될 가능성이 높다. 따라서 일반적으로 트리의 최대 깊이를 제한하여 트리를 가지치기함.

### 3.6.1 정보 이득 최대화 : 자원을 최대로 활용

정보가 풍부한 특성으로 노드를 나누기 위해 트리 알고리즘으로 최적화할 목적 함수를 정의. 

정의한 목적함수는 각 분할에서 정보 이득 최대화.

**불순도 기준을 시각적으로 비교**

클래스 1의 확률번위 [0, 1]에 대한 불순도 인덱스 시각화.

지니 불순도가 엔트로피와 분류 오차의 중간임을 관찰하기 위한 스케일 조정된 엔트로피(entropy / 2) 추가

```python
import matplotlib.pyplot as plt
import numpy as np

def gini(p):
  return (p) * (1-(p)) + (1-p)*(1-(1-p))

def entropy(p):
  return -p*np.log2(p) - (1-p)*np.log2((1-p))

def error(p):
  return 1 - np.max([p, 1-p])
```

```python
x = np.arange(0.0, 1.0, 0.01)
ent = [entropy(p) if p != 0 else None for p in x]
sc_ent = [e * 0.5 if e else None for e in ent]
err = [error(i) for i in x]
fig = plt.figure()
ax = plt.subplot(111)
for i, lab, ls, c, in zip([ent, sc_ent, gini(x), err],
                          ['Entropy', 'Entropy (scaled)',
                           'Gini Impurity',
                           'Misclassification Error'],
                          ['-', '-', '--', '-.'],
                          ['black', 'lightgray',
                           'red', 'green', 'cyan']):
  line = ax.plot(x, i, label=lab,
                 linestyle=ls, lw=2, color=c)
ax.legend(loc='upper left', bbox_to_anchor=(0.5, 1.15),
          ncol=5, fancybox=True, shadow=False)
ax.axhline(y=0.5, linewidth=1, color='k', linestyle='--')
ax.axhline(y=1.0, linewidth=1, color='k', linestyle='--')
plt.ylim([0, 1.1])
plt.xlabel('p(i=1)')
plt.ylabel('Impurity Index')
plt.show()
```

### 3.6.3 랜덤 포레스트로 여러 개의 결정 트리 연결

<랜덤 포레스트의 알고리즘>
1. n개의 랜덤한 부트스트랩 샘플을 뽑는다.
2. 부트스트랩 샘플에서 결정 트리 학습. 각 노드에서 중복을 허용하지 않고 랜덤하게 d개의 특성 선택 및 정보 이득과 같은 목적 함수를 기준으로 최선의 분할을 ㅁ나드는 특성을 사용해 노드 분할
3. 단계 1과 2를 k번 반복
4. 각 트리의 예측을 모아 다수결 투표로 클래스 레이블 할당. 

**랜덤 포레스트가 학습한 붓꽃 데이터셋의 결정 경계**

```python
from sklearn.ensemble import RandomForestClassifier

forest = RandomForestClassifier(criterion='gini',
                                n_estimators=25,
                                random_state=1,
                                n_jobs=2)
forest.fit(X_train, y_train)
plot_decision_regions(X_combined, y_combined,
                      classifier=forest, test_idx=range(105, 150))
plt.xlabel('petal length')
plt.ylabel('petal width')
plt.legend(loc='upper left')
plt.tight_layout()
plt.show()
```

25개의 결정 트리를 사용해 랜덤 포레스트 훈련. 노드를 분할하는 불순도 지표는 지니 불순도 사용.

n_jobs 매개변수를 사용해 컴퓨터의 멀티 코어를 사용해 모델 훈련 병렬화(여기서는 두 개의 코어 사용)

## 3.7 k-최근접 이웃 : 게으른 학습 알고리즘

KNN 알고리즘은 훈련 데이터에서 판별 함수를 학습하는 대신 훈련 데이터셋을 메모리에 저장하기 때문에 게으른 학습기라고 함. 

<KNN 알고리즘의 단계>
1. 숫자 k와 거리 측정 기준 선택
2. 분류하려는 샘플에서 k개의 최근접 이웃 찾기
3. 다수결 투표를 통해 클래스 레이블 할당

#### 유클리디안 거리 측정 방식을 사용한 사이킷런의 KNN 모델

**k-최근접 이웃 모델이 학습한 붓꽃 데이터셋의 결정 경계**

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=5, p=2,
                           metric='minkowski')
knn.fit(X_train_std, y_train)
plot_decision_regions(X_combined_std, y_combined,
                      classifier=knn, test_idx=range(105, 150))
plt.xlabel('petal length [standardized]')
plt.ylabel('petal width [standardized]')
plt.legend(loc='upper left')
plt.tight_layout()
plt.show()
```

KNN 모델에 다섯 개의 이웃을 지정했기 때문에 비교적 부드러운 결정 정계가 나타남.

다수결 투표가 동일할 경우 사이킷런의 KNN 구현은 샘플에 더 가까운 이웃을 예측으로 선택한다. 따라서 이웃들의 거리가 같다면 훈련 데이터셋에서 먼저 나타난 샘플의 클래스 레이블을 선택한다.
