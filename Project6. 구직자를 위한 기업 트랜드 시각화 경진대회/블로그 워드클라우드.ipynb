```python
import pandas as pd

df = pd.read_csv('스타트업 취업.csv')
df
```

```python
import numpy as np

contents = str(np.array(df['내용'].tolist()))
contents
```

```python
from konlpy.tag import Okt

t = Okt()

tokens_const = t.nouns(contents)

stop_words = ['통해', '때문', '경우', '대한', '관련', '준비', '과정', '생각', '진행', '사람', '문제', '시간', '가장',
              '선택', '분야', '시작', '부분', '방법', '영역', '위해', '학년', '여러분', '대해', '내용', '또한', '가지', '모두',
              '영진전문대', '해당', '이번', '다른', '정말', '추천', '오늘', '유형', '하나', '사업', '제공', '바로', '계열',
              '이상', '고민', '지금', '별로', '기초', '확인', '선생님', '보고', '정도', '설명', '직접', '구성', '더욱',
              '한번', '무엇', '추리', '포스팅', '미리', '인력', '우리', '할인', '실제', '기준', '양성', '접근', '신청', '대부분',
              '따라서', '이제', '기본', '현장', '사실', '하림', '피치', '먼저', '최근', '가스톤루', '방식', '여러', '다음',
              '기간', '코칭', '과목', '현재', '역시', '여기', '요즘', '단계', '문의', '걱정', '무료', '아래', '주문', '조금',
              '모든', '대기업', '취업', '기업', '지원', '중견', '참여', '개월', '최대', '스타트업', '부산']
tokens_const = [token for token in tokens_const if len(token) > 1 and token not in stop_words]
```

```python
const_cnt = {}
for word in tokens_const:
    const_cnt[word] = const_cnt.get(word, 0) + 1

sort_cnt = sorted(const_cnt.items(), key=lambda k:k[1], reverse=True)
sort_cnt
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud

font_path = 'C:/Users/홍연정/Downloads/nanum-all/나눔 글꼴/115.나눔바른펜/NanumFontSetup_TTF_BARUNPEN/NanumBarunpenB.ttf'

wordcloud = WordCloud(
    font_path = font_path,
    width=800,
    max_font_size=150,
    height=400,
    background_color='white',
    max_words=100)

wordcloud.generate_from_frequencies(const_cnt)

plt.axis('off')
plt.imshow(wordcloud, interpolation='bilinear')
plt.show()
```
