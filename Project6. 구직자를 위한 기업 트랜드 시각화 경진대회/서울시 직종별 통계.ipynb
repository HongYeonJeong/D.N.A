```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc
font_name = font_manager.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()
rc('font', family=font_name)
%matplotlib inline
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

## 2013-2019년

```python
df1 = pd.read_csv('서울시 직종별 통계_구인인원.csv')
df2 = pd.read_csv('서울시 직종별 통계_채용인원.csv')
```

```python
df1['2013'] = df1['2013'].str.replace(',', '').astype('int64')
df1['2014'] = df1['2014'].str.replace(',', '').astype('int64')
df1['2015'] = df1['2015'].str.replace(',', '').astype('int64')
df1['2016'] = df1['2016'].str.replace(',', '').astype('int64')
df1['2017'] = df1['2017'].str.replace(',', '').astype('int64')
df1['2018'] = df1['2018'].str.replace(',', '').astype('int64')
df1['2019'] = df1['2019'].str.replace(',', '').astype('int64')
```

```python
df2['2013'] = df2['2013'].str.replace(',', '').astype('int64')
df2['2014'] = df2['2014'].str.replace(',', '').astype('int64')
df2['2015'] = df2['2015'].str.replace(',', '').astype('int64')
df2['2016'] = df2['2016'].str.replace(',', '').astype('int64')
df2['2017'] = df2['2017'].str.replace(',', '').astype('int64')
df2['2018'] = df2['2018'].str.replace(',', '').astype('int64')
df2['2019'] = df2['2019'].str.replace(',', '').astype('int64')
```

```python
df1_t = df1.transpose()
df1_t.rename(columns=df1_t.iloc[0], inplace=True)
df1_t = df1_t.drop(df1_t.index[0])

df2_t = df2.transpose()
df2_t.rename(columns=df2_t.iloc[0], inplace=True)
df2_t = df2_t.drop(df2_t.index[0])
```

```python
fig, ax1 = plt.subplots(figsize=(20, 11))
line1 = ax1.plot(df1_t['합계'], marker='o',markersize=10, markeredgecolor='white', markeredgewidth=3, color='blue', label='구인인원')
line2 = ax1.plot(df2_t['합계'], marker='o',markersize=10, markeredgecolor='white', markeredgewidth=3, color='red', label='채용인원')
lines = line1 + line2
labels = [l.get_label() for l in lines]

plt.title('연도별 구인인원과 채용인원', fontsize=20)
plt.xticks(fontsize=13)
plt.xlabel('연도', fontsize=15)
plt.yticks(fontsize=13)
plt.ylabel('인원수',  fontsize=15)
plt.legend(fontsize=13)
plt.show()
```

```python
lst = ['경영 회계 사무관련직', '금융 보험관련직', '보건 의료관련직', '기계관련직', '화학관련직', '전기전자관련직','정보통신관련직']
df = df2.loc[df2['연도'].isin(lst)]
df = df.drop(df.index[0])
df = df.transpose()
df.rename(columns=df.iloc[0], inplace=True)
df = df.drop(df.index[0])

fig, ax1 = plt.subplots(figsize=(20, 11))
line1 = ax1.plot(df['금융 보험관련직'], marker='o', markersize=5, markeredgewidth=2, color='blue', label='금융 보험관련직')
line2 = ax1.plot(df['보건 의료관련직'], marker='o', markersize=5, markeredgewidth=2, color='green', label='보건 의료관련직')
line3 = ax1.plot(df['기계관련직'], marker='o', markersize=5, markeredgewidth=2, color='hotpink', label='기계관련직')
line4 = ax1.plot(df['화학관련직'], marker='o', markersize=5, markeredgewidth=2, color='purple', label='화학관련직')
line5 = ax1.plot(df['전기전자관련직'], marker='o', markersize=5, markeredgewidth=2, color='orange', label='전기전자관련직')
line6 = ax1.plot(df['정보통신관련직'], marker='o', markersize=5, markeredgewidth=2, color='red', label='정보통신관련직')

lines = line1 + line2 + line3 + line4 + line5 + line6
labels = [l.get_label() for l in lines]

plt.title('연도 및 직종별 채용인원', fontsize=20)
plt.xticks(fontsize=13)
plt.yticks(fontsize=13)
plt.xlabel('연도', fontsize=15)
plt.ylabel('채용인원', fontsize=15)
plt.legend(fontsize=14)
plt.show()
```

```python
df1_it = df1.loc[df1['연도'] == '정보통신관련직']
df1_it = df1_it.transpose()
df1_it.rename(columns=df1_it.iloc[0], inplace=True)
df1_it = df1_it.drop(df1_it.index[0])

df2_it = df2.loc[df2['연도'] == '정보통신관련직']
df2_it = df2_it.transpose()
df2_it.rename(columns=df2_it.iloc[0], inplace=True)
df2_it = df2_it.drop(df2_it.index[0])

fig, ax1 = plt.subplots(figsize=(20, 11))
line1 = ax1.plot(df1_it['정보통신관련직'], marker='o', markersize=5, markeredgewidth=2, color='blue', label='구직인원')
line2 = ax1.plot(df2_it['정보통신관련직'], marker='o', markersize=5, markeredgewidth=2, color='red', label='채용인원')

lines = line1 + line2
labels = [l.get_label() for l in lines]

plt.title('정보통신관련직 구직인원과 채용인원', fontsize=20)
plt.xticks(fontsize=13)
plt.yticks(fontsize=13)
plt.xlabel('연도', fontsize=15)
plt.ylabel('인원수', fontsize=15)
plt.legend(fontsize=15)
plt.show()
```
